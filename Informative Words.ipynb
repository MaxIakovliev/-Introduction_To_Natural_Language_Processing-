```python
import nltk
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
alice=nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
alice_fq=nltk.FreqDist(alice)
```

```python
alice_fq_100=alice_fq.most_common(100)
```

```python
moby=nltk.corpus.gutenberg.words('melville-moby_dick.txt')
moby_fq=nltk.FreqDist(moby)
moby_fq_100=moby_fq.most_common(100)
```

```python
moby_fq_100
```

```python
alice_100=[word[0] for word in alice_fq_100]
```

```python
alice_100
```

```python
moby_100=[word[0] for word in moby_fq_100]
```

```python
moby_100
```

```python
unique=set(alice_100)-set(moby_100)
```
